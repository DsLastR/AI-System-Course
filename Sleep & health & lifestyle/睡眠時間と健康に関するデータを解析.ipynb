```python
# 必要なモジュールをインポートする
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Seaborn データの可視化ライブラリ

%matplotlib inline
```

**データを読み込む**

```python
# train.csvを読み込みDataFrameを生成。
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv', encoding='shift-jis')

# 先頭を表示
df.head(20)
```

**列を組み合わせて確認**

**睡眠質列と睡眠時間列を組み合わせた集計**

```python
# 睡眠質列と睡眠時間列を組み合わせた集計
sleep_quality_duration_df = df.groupby(['Quality of Sleep', 'Sleep Duration'])['Person ID'].count().reset_index()
sleep_quality_duration_df = sleep_quality_duration_df.rename(columns={'Person ID': 'Count'})

# 集計結果を表示
print(sleep_quality_duration_df)

# ヒートマップを作成
pivot_table = sleep_quality_duration_df.pivot(index='Quality of Sleep', columns='Sleep Duration', values='Count')
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt='0f', cmap='viridis')
plt.title('睡眠質と睡眠時間の組み合わせ')
plt.xlabel('睡眠時間')
plt.ylabel('睡眠質')
plt.show()
```

**年齢列と睡眠時間列を組み合わせた集計**

```python
# 年齢と睡眠時間の関係を調べる
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Sleep Duration', data=df)
plt.title('年齢と睡眠時間的关系')
plt.xlabel('年齢')
plt.ylabel('睡眠時間')
plt.show()

# 年齢と睡眠時間との相関を調べる
correlation = df['Age'].corr(df['Sleep Duration'])
print(f"年齢と睡眠時間の相関係数: {correlation}")

# 年齢層別の平均睡眠時間を算出
age_sleep_mean = df.groupby('Age')['Sleep Duration'].mean()
print(age_sleep_mean)

# 年齢層別の平均睡眠時間を棒グラフで表示
plt.figure(figsize=(12, 6))
plt.bar(age_sleep_mean.index, age_sleep_mean.values)
plt.xlabel('年齢')
plt.ylabel('平均睡眠時間')
plt.title('年齢層別の平均睡眠時間')
plt.show()
```
